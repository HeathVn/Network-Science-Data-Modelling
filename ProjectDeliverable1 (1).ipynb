## Dataset:

Soccer Player Statistics https://www.kaggle.com/antoinekrajnc/soccer-players-statistics


## Use case:

Soccer Team Management: Use graph data to target best players at a specific position in terms of player versatility and work rate


## Data Graph Model Questions:


<ul>
  <li>Which club produces the highest value players?</li>
  <li>Who is the youngest high rating player for a particular club?</li>
  <li>Who is the oldest high rating player for a particular club with a contract that will soon expire?</li>
  <li>Who is the oldest high rating player for a particular position with a contract that will soon expire?</li>  
  <li>Is player A at position A on the same team as player B at position B?</li>
  <li>Is the best player for most teams left or right foot dominant?</li>
  <li>What work rate classification does a particular player have?</li>
</ul>

** Rating Group is either High (player ratings 85 & above), Middle( player ratings 70 - 85) or Low (player ratings 69 & below) *
