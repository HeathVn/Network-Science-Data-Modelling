```python
import neo4j
import pandas as pd
```

```python
df = pd.read_csv('FullData.csv')
print(df.shape)
df.head(3)
```

```python
df.rename({'National_Kit': 'National_Team_Kit'}, axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df.Height = df.Height.str.split(' ', expand=True)[0]
df[['Height']].head()
```

```python
df.Weight = df.Weight.str.split(' ', expand=True)[0]
df[['Weight']].head()
```

```python

club_df = df[['Club']]

club_df.drop_duplicates(subset ="Club", keep = 'first', inplace = True)
club_df.head(50)
```

```python
position_df = df[['Club','Club_Position']]
df.loc[df['Club'] == 'Free agent', 'Club'] = 'Free Agents'
df.loc[df['Club'] == 'Free Agents', 'Club_Position'] = 'Res'
position_df.head(300)
```

```python
df.insert(len(df.columns), 'Rating_Group', '')
df.head()
```

```python
df.loc[:,["Rating"]]
```

```python
df.loc[df['Rating'] >= 85, 'Rating_Group'] = 'High'
df.loc[df['Rating'] <= 84, 'Rating_Group'] = 'Medium'
df.loc[df['Rating'] <= 60, 'Rating_Group'] = 'Low'
```

```python
df.head(100)
```

```python
df.loc[:,["Rating","Rating_Group"]]
```

```python
rating_group_df = df[['Club_Position','Rating_Group']]
rating_group_df.head(100)
```

```python
player_rating_df = df[['Rating_Group','Name','Rating']]
player_rating_df.head(100)
```

```python
pd.options.display.float_format = "{:,.0f}".format
player_df = df[['Club','Name','Rating','Age','Height','Weight','Birth_Date','Preffered_Position','Nationality','National_Position','Club_Joining','Contract_Expiry','Preffered_Foot']]
player_df.head(100)
```

```python
work_rate_df = df[['Name','Work_Rate']]
work_rate_df.head(100)
```

```python
df.insert(len(df.columns), 'Physical', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Aggression','Reactions','Stamina','Strength','Jumping','Composure']
df['Physical'] = df[cols].astype(float).mean(axis=1)
```

```python
df.head()
```

```python
physical_df = df[['Name','Physical','Aggression','Reactions','Stamina','Strength','Jumping','Composure']]
physical_df.head(100)
```

```python
df.insert(len(df.columns), 'Passing', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Vision','Crossing','Short_Pass','Long_Pass','Curve','Volleys']
df['Passing'] = df[cols].astype(float).mean(axis=1)
```

```python
df.head()
```

```python
passing_df = df[['Name','Passing','Vision','Crossing','Short_Pass','Long_Pass','Curve','Volleys']]
passing_df.head(100)
```

```python
df.insert(len(df.columns), 'Defending', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Sliding_Tackle','Standing_Tackle','Interceptions']
df['Defending'] = df[cols].astype(float).mean(axis=1)
```

```python
defending_df = df[['Name','Defending','Sliding_Tackle','Standing_Tackle','Interceptions']]
defending_df.head(100)
```

```python
df.insert(len(df.columns), 'Shooting', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Heading','Weak_foot','Shot_Power','Finishing','Long_Shots','Freekick_Accuracy','Penalties']
df['Shooting'] = df[cols].astype(float).mean(axis=1)
```

```python
shooting_df = df[['Name','Shooting','Heading','Weak_foot','Shot_Power','Finishing','Long_Shots','Freekick_Accuracy','Penalties']]
shooting_df.head(100)
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Skill_Moves','Ball_Control','Attacking_Position','Balance','Agility']
df['Dribbling'] = df[cols].astype(float).mean(axis=1)
```

```python
dribbling_df = df[['Name','Dribbling','Skill_Moves','Ball_Control','Attacking_Position','Balance','Agility']]
dribbling_df.head(100)
```

```python
df.insert(len(df.columns), 'Pace', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['Acceleration','Speed']
df['Pace'] = df[cols].astype(float).mean(axis=1)
```

```python
pace_df = df[['Name','Pace','Acceleration','Speed']]
pace_df.head(100)
```

```python
df.insert(len(df.columns), 'GoalKeeping', '')
df.head()
```

```python
pd.options.display.float_format = "{:,.0f}".format
cols = ['GK_Positioning','GK_Diving','GK_Kicking','GK_Handling','GK_Reflexes']
df['GoalKeeping'] = df[cols].astype(float).mean(axis=1)
```

```python
gk_df = df[['Name','GoalKeeping','GK_Positioning','GK_Diving','GK_Kicking','GK_Handling','GK_Reflexes']]
gk_df.head(100)
```

```python
club_df.to_csv('data/club.csv', index=False)
position_df.to_csv('data/position.csv', index=False)
rating_group_df.to_csv('data/ratingGroup.csv', index=False)
player_rating_df.to_csv('data/playerRating.csv', index=False)
player_df.to_csv('data/player.csv', index=False)
work_rate_df.to_csv('data/workRate.csv', index=False)
physical_df.to_csv('data/physical.csv', index=False)
passing_df.to_csv('data/passing.csv', index=False)
defending_df.to_csv('data/defending.csv', index=False)
shooting_df.to_csv('data/shooting.csv', index=False)
dribbling_df.to_csv('data/dribbling.csv', index=False)
pace_df.to_csv('data/pace.csv', index=False)
gk_df.to_csv('data/gk.csv', index=False)
```
